```python
import json
import pandas as pd
from pathlib import Path
```

```python
# Path ke file JSON hasil generate QA
json_path = Path("output/qa_cleaning_peraturan_2023.json")

# Load JSON
with json_path.open("r", encoding="utf-8") as f:
    data = json.load(f)

print("Total QA:", len(data["qa_pairs"]))
```

```python
df = pd.DataFrame(data["qa_pairs"])
df.head()
```

```python
# Output CSV
csv_path = Path("output_csv/qa_cleaning_peraturan_2023.csv")

df.to_csv(
    csv_path,
    index=False,
    encoding="utf-8-sig" 
)
print("✅ CSV berhasil dibuat:", csv_path)
```
